# Find out the monthly revenue and analyze the findings

```python
data['month_year'] = data['Order Date'].apply(lambda x: x.strftime('%Y-%m'))
df3 = data['month_year'] 
print(df3)
```

```python
df4 = data.groupby('month_year').sum()['Sales'].reset_index()
print(df4)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,6))
plt.plot(df4['month_year'], df4['Sales'])
plt.xticks(rotation='vertical', size=8)
plt.xlabel("Year_Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Almost every year, there is a decrease in sales from January to February, followed by a steep increase in March. There are steep fluctuations from August to December.
# Overall, the sales of the Superstore have grown from 2014 to 2017.
df4.describe()
```

# Find out the yearly revenue and analyze the findings

```python
data['year'] = data['Order Date'].apply(lambda x: x.strftime('%Y'))
df5 = data['year'] 
print(df5)
```

```python
df6 = data.groupby('year').sum()['Sales'].reset_index()
print(df6)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df6['year'], df6['Sales'])
plt.xticks(rotation='vertical', size=8)
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
# The sales decreased from 2014 to 2015, and then increased from 2015 to 2017
df6.describe()
```

# Finding out the monthly growth rate and analyse the findings

```python
df4['Last_Month'] = df4['Sales'].shift()
df4['Growth'] = (df4['Sales']/df4['Last_Month'])-1
print(df4['Growth'])
```

```python
df4
```

```python
plt.figure(figsize=(15,8))
plt.plot(df4['month_year'], df4['Growth'])
plt.xticks(rotation='vertical', size=8)
plt.xlabel("Year_Month")
plt.ylabel("Growth")
plt.show()
```

```python
df4['Growth'].max()
# Maximum Monthly growth rate is in the year 2014, in March
```

```python
df4.describe()
```
